#Q-learning 

##Q-learning Method

When we train our agent using Q-learning method, the agent takes action under certain policy and gets series observations from the game, which is a path shown as:

\begin{equation*}
\{ s_0;a_1,r_1,s_1;a_2,r_2,s_2;.....;a_n,r_n,s_n;\}\ 
\end{equation*}


If the state space is discrete and the transition is known, for each (s, a, r, s’) tuple, the Q value of each action can be updated using the Bellman equation:


\begin{equation*}
Q_\pi (s,a) = \sum_{s^{\prime}} T (s,a,s^{\prime}) [Reward(s,a,s^{\prime})+ \gamma \max_{a^{\prime}}Q_\pi(s^{\prime},a^{\prime})]
\end{equation*}


When transitions are unknown, we can directly update Q using every single observation tuple (s, a, r, s’ ):


\begin{equation*}
Q_opt(s,a) \leftarrow Q_\pi (s,a) - \eta[Q_opt(s,a)−u] 
                            = Q_opt(s,a) - \eta[Q_opt(s,a) - (r+\max_{a^{\prime}}Q_opt(s^{\prime},a^{\prime}))]
\end{equation*}


Due to the high-dimensional image input, the state will be really large and impossible to learn. So we use a Convolutional Neural Network to functionally approximate the Q value of each state. Assuming the weights of CNN is w, the learning process is then equivalent to optimize the cost for all observations:


\begin{equation*}
\min_{w} \sum_{(s,a,r,s^{\prime})} (Q_opt(s,a)-(r+\max_{a^{\prime}}Q_opt(s^{\prime},a^{\prime})))^2
\end{equation*}


The optimization can be done using the backward propagation training method of CNN.

#Approach Taken

The size of our state space immediately rules out several categories of RL algorithms:

• Maximum likelihood-based methods are out of the question since at any given timestep
it is overwhelmingly likely that we have never before observed precisely the same
screen, meaning that the vast majority of our T and R estimates would be undefined.

• Table-based model-free methods such as Q-learning are also ruled out, as
the space of state-action pairs (s, a) is far too large to be stored in tabular form, even
given all the memory in the world. Using sparse matrix representations for Q-tables
is no help since it does not address the underlying issue that the huge majority of
Q(s, a) values will never see an update.

Hence, we decided to go ahead with  Deep Q-learning for this problem.

The action space of our model is that the agent can go “left, right, straight, brake”. The discount factor γ is chosen to be 0.99 to make the Q converge faster.


##Algorithm description

The essence of deep Q-learning is the estimation of Q*(s, a) using a type of neural networkcalled a Deep Q-Network (DQN) parametrized by a vector θ. At training iteration i we write the network parameters as θi. The loss at this step is given by the temporal difference error and differentiating with respect to θi, we find that loss is minimized when the Bellman equation is satisfied,


\begin{equation*}
Q (s;a;θ_i) = E_{s^{\prime}} [R(s, a) +\gamma \max_{a^{\prime}}Q_\pi(s^{\prime},a^{\prime};θ_i-1) ]
\end{equation*}

This in conjunction with the fact that neural networks are universal function approximators
implies that, given sufficient training data, a DQN will learn the optimal values of Q.

Deep Q-learning needs some tweaks to succeed in gaming contexts. The first involves
decorrelating successive training samples. We do this using experience replay. With experience replay all “experiences” (s, a, r, s') observed in training are stored in a database D,
and at each iteration a random sample of experiences is drawn to train the network. For
simplicity we trained on only one random experience in D at each timestep. The second
addresses the issue of a target value which changes in time with the network parameters θi ,using an auxiliary network called a target network.

###Training

If we train our CNN at every single frame, the training may suﬀer from several problems. Since the state at time t is highly correlated with state at time t+1, gradient descent after consecutive steps will cause erratic updates making the training very slow. In order to ﬁx this problem and make the training faster as well as more stable, we use the batch training method. During training, we maintain a memory with capacity 50000 and save the observations at each frame. Then after the observation phase we would choose a batch of data from the memory randomly to train our CNN. This experience replay strategy makes the training much more eﬃcient and the experience observed are no longer highly correlated

**Training Parameters**

If we choose the optimal action every time, the Q-learning training may converge to a local optimal policy. In order to make the agent explore more states and policies, we use the ε-greedy method to do random action under some probability. The value of decreases linearly overtime. The choice of initial ε and the total steps of exploring are essential for our training because our game is quite diﬀerent from other games.

## Value network and target network
 We used our neural network to estimate the value of our next state, and used that estimate to update our network again. This can sometimes lead to "unstable" situations, where your network makes a strange decision which keeps going through your network each time you update it. 

This is why we will use a value and a target network here... It boils down on constructing two neural networks with the same layer. To update our target network we use our value network like this: 

> $Qtarget(state,action) = reward + gamma * Qvalue(nextstate, max_action)$

While we train the target network gets better and better at estimating the value of a state. We copy the weights of our target network to the value network every X iterations, where you can define how often this happens. 

We also decide to not train every frame, but every X frames so you can put more playtime in training. 

```python
TARGET_UPDATE_INTERVAL = int(Latest_Parameters_used["TARGET_UPDATE_INTERVAL"])   # The frequency with which the target network is updated
TRAIN_INTERVAL = int(Latest_Parameters_used["TRAIN_INTERVAL"])  # The agent selects X actions between successive updates
```

## Defining Network


```python
def build_network(name):
    with tf.variable_scope(name):
        state_placeholder = tf.placeholder(tf.float32, [None, FRAME_WIDTH, FRAME_HEIGHT, STATE_HISTORY])
        
        #conv1 = tf.layers.conv2d(state_placeholder, 32, (8,8), strides=(4,4), activation=tf.nn.relu, name="firstconvlayer")
        conv1 = tf.layers.conv2d(state_placeholder, 8, (7,7), strides=(3,3), activation=tf.nn.relu, name="firstconvlayer")
        maxp1 = tf.layers.max_pooling2d(conv1, (2,2),(2,2), name='maxp1')
        
        #conv2 = tf.layers.conv2d(conv1, 64, (4,4), strides=(2,2), activation=tf.nn.relu, name="secondconvlayer")
        conv2 = tf.layers.conv2d(maxp1, 16, (3,3), strides=(2,2), activation=tf.nn.relu, name="secondconvlayer")
        maxp2 = tf.layers.max_pooling2d(conv2, (2,2),(2,2), name='maxp2')
        
        #conv3 = tf.layers.conv2d(conv2, 64, (3,3), strides=(1,1), activation=tf.nn.relu, name="thirdconvlayer")
        flattened = tf.contrib.layers.flatten(maxp2)

        #dense1 = tf.layers.dense(flattened, 512, activation=tf.nn.relu, name="firstdenselayer")
        dense1 = tf.layers.dense(flattened, 256, activation=tf.nn.relu, name="firstdenselayer")
        
        #dense2 = tf.layers.dense(dense1, 256, activation=tf.nn.relu, name="Seconddenselayer")
        
        #q_values = tf.layers.dense(dense2, num_actions, name="q_valueslayer")
        q_values = tf.layers.dense(dense1, num_actions, name="q_valueslayer")
        
        action_distribution = tf.nn.softmax(q_values)

    
    ## Get a collection of all trainable parameters in this scope so we can build an update function for the value and target network
    trainable_parameters = [v for v in tf.get_collection(tf.GraphKeys.TRAINABLE_VARIABLES) if v.name.startswith(name)]
    
    return state_placeholder, q_values, trainable_parameters, action_distribution
```

## Loss function
As loss function you can use mean-squared error, but note that this error metric can give us HUGE values if our network makes a large mistake. An error measure that is a bit "smoother" for larger errors is the Huber loss. This loss is quadratic for small errors (between -1 and 1), and linear after that. An image of this can be seen below: 

![huber](https://upload.wikimedia.org/wikipedia/commons/thumb/c/cc/Huber_loss.svg/300px-Huber_loss.svg.png)

The blue line indicates the mean squared error gradients, while the green line indicates the Huber loss. For small errors the lines overlap. 

```python
# Clip the error, the loss is quadratic when the error is in (-1, 1), and linear outside of that region
with tf.variable_scope("huber_loss"):
    error = tf.abs(y - q_value)
    quadratic_part = tf.clip_by_value(error, 0.0, 1.0)
    linear_part = error - quadratic_part
    #loss = tf.reduce_mean(0.5 * tf.square(quadratic_part) + linear_part)
```

## The optimizer and Loss


```python
LEARNING_RATE = float(Latest_Parameters_used["LEARNING_RATE"])  # Learning rate used by RMSProp
MOMENTUM = float(Latest_Parameters_used["MOMENTUM"])   # Momentum used by RMSProp
MIN_GRAD = float(Latest_Parameters_used["MIN_GRAD"])   # Constant added to the squared gradient in the denominator of the RMSProp update


#  HUBER Loss parameters ###
#optimizer = tf.train.RMSPropOptimizer(LEARNING_RATE, momentum=MOMENTUM, epsilon=MIN_GRAD)
#grads_update = optimizer.minimize(loss, var_list=q_network_weights)


loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(labels = a, logits = q_values))
optimizer = tf.train.AdamOptimizer(LEARNING_RATE)
grads_update = optimizer.minimize(loss, var_list=q_network_weights)
```

## How do we determine our action? 
At the start of our program we always want to select a random action. This is determined by the INITIAL_REPLAY_SIZE variable defined above. If we are passed this amount we draw a number, check if it's bigger than epsilon. If not: we select a random action, else we evaluate our state and select the best action!

```python
# Convert action to one hot vector used during explore
import pandas as pd
def int_to_onehot(n, n_classes):
    v = [0] * n_classes
    v[n] = 1
    return v

def onehot_to_int(v):
    return v.index(1)
```

```python

def get_action(state, epsilon, exploit, explore,last_50_reward_value):
    rr = random.random()
   # print(" random.random():",  rr)
    if ((last_50_reward_value==True) and (t > INITIAL_REPLAY_SIZE)):
       # print("inside True")
        action = random.randrange(num_actions)
        action = np.array(int_to_onehot(action, num_actions))
        exploit+=1      
    elif epsilon >= rr or t < INITIAL_REPLAY_SIZE:
        action = random.randrange(num_actions)
        action = np.array(int_to_onehot(action, num_actions))
        exploit+=1
   #     print("inside action", "random:", rr, "epsilon:", epsilon)
    else:
        state = np.float32(state / 255.0)
        action = sess.run(action_distribution,feed_dict={s: state.reshape([1, FRAME_WIDTH, FRAME_HEIGHT, STATE_HISTORY])/ 255.0})
        explore+=1
        
        action = action.flatten()

    # Anneal epsilon linearly over time
    if epsilon > FINAL_EPSILON and t >= INITIAL_REPLAY_SIZE:
        epsilon_before=epsilon
        epsilon -= epsilon_step
  #      print("epsilon before:", epsilon_before, "epsilon after:", epsilon)

    return action, exploit, explore, epsilon
```

## Training a network
Time to train our network. We pass the replay memory to a train_network function and feed the states, actions, and the estimated values by the target network to our normal network!

```python
BATCH_SIZE = int(Latest_Parameters_used["BATCH_SIZE"]) 

def train_network(replay_memory):
    state_batch = []
    action_batch = []
    reward_batch = []
    next_state_batch = []
    terminal_batch = []
    y_batch = []

    # Sample random minibatch of transition from replay memory
    minibatch = random.sample(replay_memory, BATCH_SIZE)
    for data in minibatch:
        state_batch.append(data[0])
        action_batch.append(data[1])
        reward_batch.append(data[2])
        next_state_batch.append(data[3])
        terminal_batch.append(data[4])

    # Convert True to 1, False to 0
    terminal_batch = np.array(terminal_batch) + 0

    target_q_values_batch = target_q_values.eval(
        feed_dict={st: np.float32(np.array(next_state_batch) / 255.0)})
    
    y_batch = reward_batch + (1 - terminal_batch) * GAMMA * np.max(target_q_values_batch, axis=1)
    
    local_loss, _ = sess.run([loss, grads_update], feed_dict={
        s: np.float32(np.array(state_batch) / 255.0),
        a: action_batch,
        y: y_batch
    })
    
    
    return local_loss
```

```python
  
max_steps =1000

start = time.time()

## Reset variables
total_reward = 0
total_q_max = 0
total_loss = 0
duration = 0

exploit=0
explore=0


Final_epsilon=[]

env.reset()

prev_screen = env.render(mode='rgb_array')
plt.imshow(prev_screen)

# Final Result arrays
Action_Final_List=[]
Final_Reward=[]
Final_Num_Steps=[]
Final_total_q_max=[]
Final_Loss=[]
Final_epsilon=[]
Final_reward_per_episode=[]

# Epoch start
for k in range(NUM_EPISODES):
    t_reward=[]
    t_avg_max_q=[]
    t_loss=[]
    ## Reset the network and get an initial state
    terminal = False
    
    observation = env.reset()
    
    for _ in range(random.randint(1, NO_OP_STEPS)):
        last_observation = observation
        observation, _, _, _ = env.step([0,0,0])  # Do nothing
        
    state = get_initial_state(observation, last_observation)

    # Before every episode reinitializing, so that we can find how many actions popped using explore/exploit.
    
    exploit=0
    explore=0
    epsilon_per_episode=[]
    reward_per_episode=[]
    last_50_reward_val=[]
    print("length of last_50_reward_val:", len(last_50_reward_val))
    
    i=0
    while not terminal:
    #for i in range(max_steps+1):
        #print("====================================================================================")
        # For the first 10 steps passing STRAIGHT action
        if i < 10:
          last_observation = observation
          #action_first, explotCNT, exploreCNT = get_action(state, epsilon,exploit,explore)
          exploit+=1
          action = possible_actions[2]
          screen = env.render(mode='rgb_array')
          observation, reward, terminal, _ = env.step(action)
          processed_observation , processed_observation_grey = preprocess(observation, last_observation)
          next_state = np.dstack((state[:, :, 1:], processed_observation))
          
          total_reward += reward
          total_q_max += np.max(q_values.eval(feed_dict={s: [np.float32(state / 255.0)]}))
          
          plt.imshow(processed_observation_grey,cmap='Greys', interpolation ='none' , vmin=0, vmax=1)
          plt.show()
          print("i:", i)
          i+=1
          t += 1
          state = next_state
                    
        else:
          
          if i > 25:
            
            # check if the :AST 50 values are all NEGATIVE or not. If ALL the negative than will EXPLOIT the action.
            last_50_reward_val = all(n < 0 for n in (np.array(reward_per_episode)[-50:]))
            
          last_observation = observation

          action_first, explotCNT, exploreCNT, epsilon_result = get_action(state, epsilon,exploit,explore,last_50_reward_val)
        
          exploit=explotCNT
          explore=exploreCNT
          epsilon=epsilon_result
          
          action_index = np.argmax(action_first.ravel())
       
          action = possible_actions[action_index]
        
          # Render Action After 100 steps and download Event file too
          #if i%10000 == 0:
          
          
          screen = env.render(mode='rgb_array')
          #plt.imshow(screen)
          #ipythondisplay.clear_output(wait=True)
          #ipythondisplay.display(plt.gcf())
                            
          observation, reward, terminal, _ = env.step(action)
          processed_observation , processed_observation_grey = preprocess(observation, last_observation)

          
          # storing Epsilon and Reward per step
          epsilon_per_episode.append(epsilon)
          reward_per_episode.append(reward)
          
          
          # Dislay Grey Scale Image
        
          if i%20000 == 0:
            plt.imshow(processed_observation_grey,cmap='Greys', interpolation ='none' , vmin=0, vmax=1)
            plt.show()
      
          next_state = np.dstack((state[:, :, 1:], processed_observation))

          # Store transition in replay memory
          replay_memory.append((state, action_first, reward, next_state, terminal))

          ## If the replay memory is too large, remove the first part of the memory
          if len(replay_memory) > NUM_REPLAY_MEMORY:
              replay_memory.popleft()
      
          
          ## If we have enough frames to start learning, train!
          if duration>1:
            pass
            #print("episode:", k, "t:", t, "action:", action, "epsilon", epsilon, "terminal", terminal,"TRAIN_INTERVAL", TRAIN_INTERVAL, "INITIAL_REPLAY_SIZE:", INITIAL_REPLAY_SIZE, "TARGET_UPDATE_INTERVAL:", TARGET_UPDATE_INTERVAL,  "total_reward:", total_reward, "total_q_max/float(duration):", total_q_max/float(duration),"duration:", duration , "episode_avg_loss:", total_loss/duration)
          if t >= INITIAL_REPLAY_SIZE:
              if t % TRAIN_INTERVAL == 0:
                  total_loss += train_network(replay_memory)

              ## Set the target network variables equal to what we trained
              if t % TARGET_UPDATE_INTERVAL == 0:
                 # print("inside t % TARGET_UPDATE_INTERVAL")
                  sess.run(update_target_network)
                
          ## Increase variables like reward and q
          total_reward += reward
          total_q_max += np.max(q_values.eval(feed_dict={s: [np.float32(state / 255.0)]}))
          
          if i%100==0:
            print("episode:", k,"i:", i, "t:", t, "action:", action, "epsilon", epsilon, "terminal", terminal, "explotCNT:", explotCNT, "exploreCNT:", exploreCNT, "TRAIN_INTERVAL", TRAIN_INTERVAL, "INITIAL_REPLAY_SIZE:", INITIAL_REPLAY_SIZE, "TARGET_UPDATE_INTERVAL:", TARGET_UPDATE_INTERVAL,  "total_reward:", total_reward, "total_q_max/float(duration):", total_q_max/float(duration),"duration:", duration , "episode_avg_loss:", total_loss/duration)
              
          
          duration += 1
          
          i+=1
          t += 1
          state = next_state
        
          BKP_duration=duration
        
          # Inorder to terminate the episode before so that not enough -ve rewards gets accumulated, i used set number of steps 
          # to terminate the episode
        
          #if i == max_steps: terminal = True
        
        
          if terminal:
           
              ## Write summary for TensorBoard
              if t >= INITIAL_REPLAY_SIZE:
                  #print("t inside write:") #, t, "total_reward:", total_reward, "total_q_max/float(duration):", total_q_max/float(duration),"duration:", duration , "episode_avg_loss:", total_loss/duration)
                  summary_str = sess.run(summary_op, feed_dict={episode_total_reward: float(total_reward), episode_avg_max_q: float(float(total_q_max)/float(duration)), episode_duration: duration , episode_avg_loss: float(float(total_loss)/duration)})
                  summary_writer.add_summary(summary_str, episode + 1)
                
                  t_reward.append(float(total_reward))
                  t_avg_max_q.append(float(float(total_q_max)/float(duration)))
                  t_loss.append(float(float(total_loss)/duration))
                
                  """
                  # Draw Plot
                  fig, (ax1, ax2,ax3) = plt.subplots(nrows=1, ncols=3,figsize=(15,5))

                  ax1.plot(t_reward)
                  ax2.plot(t_avg_max_q)
                  ax3.plot(t_loss)

                  # Display the figure
                  plt.show()
                  """
               
              ## Debug
              print("episode:", k, "t:", t,"t_reward:", t_reward, "action:", action, "epsilon",epsilon, "explotCNT:", explotCNT, "exploreCNT:", exploreCNT, "terminal:", terminal, "Finished episode %d/%d which took %d frames" % (episode+1, NUM_EPISODES, duration))
              
           
              ## Reset variables
              total_reward = 0
              total_q_max = 0
              total_loss = 0
              duration = 0
              episode += 1
            
              break

          if t %10000 == 0:
              saver.save(sess,SAVE_SUMMARY_PATH+'Model_22') 
            
    exploit=explotCNT
    explore=exploreCNT
    Action_list=[k,exploit,explore]
    
    Action_Final_List.append(Action_list)
    Final_Reward.append(t_reward)
    Final_total_q_max.append(t_avg_max_q)
    Final_Num_Steps.append(BKP_duration)
    Final_Loss.append(t_loss)
    Final_epsilon.append(epsilon_per_episode)
    Final_reward_per_episode.append(reward_per_episode)
    
end = time.time()
time_taken = end - start

#print(Action_Final_List)    
saver.save(sess,SAVE_SUMMARY_PATH+'Model_22')

ipythondisplay.clear_output(wait=True)
env.close()

```
